```python

from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import faiss

embedding_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# Define the summarization tokenizer and model (used for generating summaries)
summarization_tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
summarization_model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")

def search_paper(query, k=5):
    query_embedding = embedding_model.encode([query])
    faiss.normalize_L2(query_embedding)
    D, I = index.search(query_embedding, k)
    return D, I

def summarizer_search(query, k=5):
    D, I = search_paper(query, k)
    for score, idx in zip(D[0], I[0]):
        paper_title = df.iloc[idx]['title']
        paper_abstract = df.iloc[idx]['abstract']

        # Generate summary for the abstract using the summarization model
        inputs = summarization_tokenizer([paper_abstract], max_length=1024, return_tensors='pt', truncation=True)
        summary_ids = summarization_model.generate(
            inputs['input_ids'],
            num_beams=4,
            max_length=150,
            min_length=30,
            early_stopping=True
        )
        generated_summary = summarization_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

        print(f"\n--- Search Result (Score: {score:.4f}) ---")
        print(f"Title: {paper_title}")
        print(f"Abstract: {paper_abstract[:500]}...")
        print(f"Summary: {generated_summary}")

summarizer_search("deep learning for medical image analysis", k=3)
```

Named Entity Recognition (NER)


```python
from transformers import pipeline

# Initialize the NER pipeline
ner_pipeline = pipeline("ner", model="dslim/bert-base-NER")
print("NER pipeline initialized successfully.")
```

```python
sample_title_for_ner = df.iloc[10466]["title"]
print("--- Sample Title ---")
print(sample_title_for_ner)

# Perform NER on the title
ner_results_title = ner_pipeline(sample_title_for_ner)

print("\n--- NER Results for Title ---")
if ner_results_title:
    for entity in ner_results_title:
        print(f"Entity: {entity['word']}, Type: {entity['entity']}, Score: {entity['score']:.4f}")
else:
    print("No named entities found in the title.")
```
